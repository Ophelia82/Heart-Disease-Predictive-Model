```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
location = 'nhanes.xlsx'
```

```python
dfmaster = pd.read_excel(location, index_col='SEQN')
dfmaster.head()
```

```python
dfglu = pd.read_excel('GLU_H_trim.xlsx', index_col='SEQN')
dfglu.head()
```

```python
final = pd.merge(dfmaster, dfglu, how='left', on='SEQN')
final.head()
```

```python
writer = pd.ExcelWriter('final_assembled.xlsx', engine = 'xlsxwriter')
final.to_excel(writer, sheet_name='names')
writer.save()
```
