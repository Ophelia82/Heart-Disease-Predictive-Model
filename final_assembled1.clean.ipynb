```python
filename = 'final_assembled.xlsx'
df = pd.read_excel(filename)
df.head()
```

```python
# drop Num NaN values
dfnum = df['Num'].dropna()
```

```python
dfnum = df.dropna(subset=['Num'])
```

```python
# drop trestbps (systolic blood pressure) with missing values
dfnum2 = dfnum.dropna(subset=['trestbps'])
dfnum2
```

```python
# check to see which columns have missing values
dfnum2.isnull().sum()
```

```python
df3 = dfnum2.drop('fbs', axis=1)
```

```python
# check to see which columns have missing values
df3.isnull().sum()
```

```python
# drop CDQ002 (chest pain when walking uphill and walking fast) with missing values
dfnum4 = df3.dropna(subset=['CDQ002'])
dfnum4
```

```python
dfnum4.isnull().sum()
```

```python
# find the mean of chol (cholesterol) by age and use mean to fill in empty values
dfnum4.groupby(['Age'])['chol'].mean()
```

```python
# find the mean of thalrest (resting heart rate) by age and use mean to fill in empty values
dfnum4.groupby(['Age'])['thalrest'].mean()
```

```python
dfnum4['thalrest'].fillna(dfnum4.groupby(['Age'])['thalrest'].transform('mean'), inplace=True)
```
